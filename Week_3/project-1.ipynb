```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox

# Load the employee dataset
try:
    tf = pd.read_csv("GIG-logistics(in).csv")  # Ensure your CSV file is in the same folder
except FileNotFoundError:
    messagebox.showerror("Error", "GIG-logistics.csv file not found.")
    raise

# Normalize data (remove extra spaces, ensure uniform capitalization)
tf["Name"] = tf["Name"].str.strip().str.title()
tf["Department"] = tf["Department"].str.strip().str.title()

# Function to verify employee
def verify_employee():
    name = name_entry.get().strip().title()
    dept = dept_entry.get().strip().title()

    if name == "" or dept == "":
        messagebox.showwarning("Input Error", "Please enter both name and department.")
        return

    # Check if employee exists
    employee_exists = ((tf["Name"] == name) & (tf["Department"] == dept)).any()

    if employee_exists:
        # Get all employees in the same department
        dept_members = tf[tf["Department"] == dept]["Name"].tolist()

        message = f"✅ Welcome, {name}! You are verified as an employee in {dept} department.\n\n"
        message += "Other team members:\n"
        message += "\n".join([member for member in dept_members if member != name]) or "No other members listed."

        messagebox.showinfo("Access Granted", message)
    else:
        messagebox.showerror("Access Denied", f"❌ {name} is not found in department '{dept}'.")

# Create the main window
root = tk.Tk()
root.title("GIG Logistics Employee Verification")
root.geometry("400x250")
root.config(bg="#f0f0f0")

# Labels and entries
tk.Label(root, text="Employee Name:", bg="#f0f0f0", font=("Arial", 10)).pack(pady=5)
name_entry = tk.Entry(root, width=40)
name_entry.pack(pady=5)

tk.Label(root, text="Department:", bg="#f0f0f0", font=("Arial", 10)).pack(pady=5)
dept_entry = tk.Entry(root, width=40)
dept_entry.pack(pady=5)

# Verify button
verify_btn = tk.Button(root, text="Verify Employee", command=verify_employee, bg="#0066cc", fg="white", width=20)
verify_btn.pack(pady=15)

# Exit button
exit_btn = tk.Button(root, text="Exit", command=root.quit, bg="#cc0000", fg="white", width=10)
exit_btn.pack(pady=5)

# Start Tkinter event loop
root.mainloop()
```

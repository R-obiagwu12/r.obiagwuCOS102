```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head(8)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Fill number columns
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Fill text columns
df['Cabin'] = df['Cabin'].fillna("Unknown")
df['Embarked'] = df['Embarked'].fillna("Unknown")
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=30)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Most passengers were young adults.
Few elderly passengers

The data is slightly right-skewed, meaning there are more younger people than older ones.

```python
plt.figure(figsize=(8,5))
plt.boxplot(df['Fare'])
plt.title("Boxplot of Fare")
plt.ylabel("Fare")
plt.show()
```

Most passengers paid low to moderate fares.

There are several outliers, indicating some passengers paid significantly more.
This means there is a large difference in ticket classes and wealth levels.

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['Fare'])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

There is no relationship between age and fare. High fares are not limited to a particular age group.

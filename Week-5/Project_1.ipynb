```python
import tkinter as tk 
import random

class Employees:
    employees = ["Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James", "Michell Taiwo", "Abraham Jones" , "Nicole Anide", "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"]
    tasks = ["Loading", "Transporting", "Reveiwing Orders", "Customer Service", "Delivering Items"]

     # Check if employee exists
    def check_employee(self, name):
        return name in self.employees

    # Take attendance
    def take_attendance(self, name):
        return f"✅ Attendance taken for {name}"

    # Assign random task
    def assign_task(self):
        return random.choice(self.tasks)

    # Refuse access
    def refuse_access(self):
        return "❌ Access Denied! You are not an employee."


# -------- GUI PROGRAM -------- #

def submit():
    name = entry.get().title()   # 👈 converts to proper format

    if emp.check_employee(name):
        attendance = emp.take_attendance(name)
        task = emp.assign_task()
        result_label.config(text=f"{attendance}\nTask: {task}")
    else:
        result_label.config(text=emp.refuse_access())
        

# Create object
emp = Employees()

# GUI window
root = tk.Tk()
root.title("Employee Verification System")
root.geometry("400x300")

# Label
label = tk.Label(root, text="Enter Employee Name:")
label.pack(pady=10)

# Entry box
entry = tk.Entry(root, width=30)
entry.pack(pady=5)

# Button
button = tk.Button(root, text="Submit", command=submit)
button.pack(pady=10)

# Result display
result_label = tk.Label(root, text="", fg="blue")
result_label.pack(pady=20)

# Run app
root.mainloop()
```

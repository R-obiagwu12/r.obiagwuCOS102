```python
import tkinter as tk
from tkinter import messagebox

# Parent Class
class Zenith:
    def mutual_services(self):
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance"
        ]

    def unique_services(self):
        return []


# Child Class for Retail Banking
class RetailBanking(Zenith):
    def unique_services(self):   # Polymorphism (method overriding)
        return [
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"
        ]


# Child Class for Global Banking
class GlobalBanking(Zenith):
    def mutual_services(self):   # Override because no mutual services listed
        return []

    def unique_services(self):
        return [
            "Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]


# Child Class for Commercial Banking
class CommercialBanking(Zenith):
    def unique_services(self):
        return [
            "Advisory services"
        ]


# Function to display services
def show_services():
    name = name_entry.get().strip().lower()
    division = division_entry.get().strip().lower()

    if name == "mary" and division == "retail":
        employee = RetailBanking()
        emp_name = "Mary"
        div_name = "Retail Banking"

    elif name == "agatha" and division == "global":
        employee = GlobalBanking()
        emp_name = "Agatha"
        div_name = "Global Banking"

    elif name == "noel" and division == "commercial":
        employee = CommercialBanking()
        emp_name = "Noel"
        div_name = "Commercial Banking"

    else:
        messagebox.showerror("Error", "Invalid employee name or division!")
        return

    services = employee.mutual_services() + employee.unique_services()

    result = f"Employee Name: {emp_name}\nDivision: {div_name}\n\nServices Rendered:\n"
    for service in services:
        result += "- " + service + "\n"

    output_label.config(text=result)


# GUI Window
root = tk.Tk()
root.title("Zenith Bank Nigeria")
root.geometry("500x500")

# Labels and Entry Boxes
tk.Label(root, text="Employee Name:").pack()
name_entry = tk.Entry(root, width=30)
name_entry.pack()

tk.Label(root, text="Division (Retail / Global / Commercial):").pack()
division_entry = tk.Entry(root, width=30)
division_entry.pack()

# Button
tk.Button(root, text="Show Services", command=show_services).pack(pady=10)

# Output Label
output_label = tk.Label(root, text="", justify="left", wraplength=450)
output_label.pack(pady=10)

# Run App
root.mainloop()
```

```python
# Program to calculate Annual Tax Revenue (ATR)

# Collect input from the user
experience = int(input("Enter years of work experience: "))
age = int(input("Enter staff age: "))

# Determine ATR 
if experience > 25 and age >= 55:
    ATR = 5600000
elif experience > 20 and age >= 45:
    ATR = 4480000
elif experience > 10 and age >= 35:
    ATR = 1500000
else:
    ATR = 550000

# Display the result
print("Annual Tax Revenue (ATR): ₦", ATR)
```

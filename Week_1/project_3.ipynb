```python
## principal number
pmt = float(input("Enter Payment per period"))
## rate number
r = float(input("Enter Rate"))
## compounds number
n = float(input("Enter Interest compounds"))
## time number
t = float(input("Enter Time"))
## application of formula
a = pmt * (((1.0 + r/n) ** (t * n) - 1) / (r/n))
print("The amount is : ", a)
## simple interest calculation
ap = a - pmt
print("The annuity plan is : ", ap)
                  
```

```python
## principal number
p = float(input("Enter Principal"))
## rate number
r = float(input("Enter Rate"))
## time number
t = float(input("Enter Time"))
## application of formula
a = p * (1.0 + r/100.0) * t
print("The amount is : ", a)
## simple interest calculation
si = a - p 
print("The simple interest is : ", si)
                  
                  
```
